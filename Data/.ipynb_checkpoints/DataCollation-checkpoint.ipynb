```python
#Dropping those with one tweet from the dataframe
tweetfinalC = tweetfinalB[~tweetfinalB['writer'].isin(abc)]
tweetfinalC
```

```python
tweetrowtotal = tweetfinalC["comment_num"] + tweetfinalC["retweet_num"] + tweetfinalC["like_num"]
tweetfinalC["total"] = tweetrowtotal
tweetfinalC
```

```python
tweetwritergroup = tweetfinalC.groupby(["writer"])
```

```python
tweettotalsumbywriter = tweetwritergroup.sum()
tweettotalsumbywriter
```

```python
tweettotalsumbywriter.sort_values("total", ascending=False)
```

```python
top100writers = tweettotalsumbywriter.sort_values("total", ascending=False).head(400)
top100writers
```

```python
bottom12writers = tweettotalsumbywriter.sort_values("total").head(12)
bottom12writers
```

```python
bottom12index = bottom12writers.index.tolist()
print(len(bottom12index))
print(bottom12index)
```

```python
#Dropping bottom 12 writers of total activity
tweetfinalD = tweetfinalC[~tweetfinalC['writer'].isin(bottom12index)]
tweetfinalD
```
